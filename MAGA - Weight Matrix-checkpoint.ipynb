```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Nx = 200
Nt = 10

t0 = 0
t1 = 1.

rand_Ics = np.sort(np.random.random(Nx))
uni_Ics = np.linspace(0., 1., Nx)
norm_Ics = np.random.normal(loc=0.5, scale=0.3, size=Nx)

random_start = False
normal_start = False
uniform_start = True

adverts = True
advertisers = [0.5, 0.1]
charisma = [10, 100]
advertiser_drift = False
drift = [-1.1, 1.8] 
advertiser_sine_wave = False
polling = False

brownian_motion = False
h = .05 # the variance of the brownian motion

gradient = 1
bias = 1

# dont worry about this -- yet to be implemented
T = np.linspace(0, Nt, Nt)
T_scaled = (T/Nt)*2*np.pi
```

```python
A = np.zeros([Nt, Nx])

if random_start:
    A[0, :] = rand_Ics
elif uniform_start:
    A[0, :] = uni_Ics
elif normal_start:
    A[0, :] = norm_Ics

ad_pos = []

if adverts:
# ------ Add advertisers into the system ---------
    for i in range(len(advertisers)):
        A_new = np.hstack((A, np.atleast_2d(np.zeros(Nt)).T))
        A_new[0, -1] = advertisers[i]
        A = A_new

    A[0, :] = np.sort(A[0, :])


    for stance in advertisers:
        pos = np.argwhere(A[0, :] == stance)[0][0]
        ad_pos.append(pos)
        
ad_pos = np.array(ad_pos)
# ------------------------------------------------

Ics2 = A[0, :]

for t in range(Nt-1):
    
    U = A[t, :]
    
    for x in range(A.shape[1]):
        
        # create weight matrix
        W = (-gradient*abs(U - U[x]) + bias)
        
        if adverts:
            if x in ad_pos:
                print(U[x])
                ad_index = np.where(ad_pos == x)[0][0]
                W[x] += charisma[ad_index]

            if x not in ad_pos:

                A[t+1, x] = (W @ A[t, :]) / np.sum(W)
                
        else:
            A[t+1, x] = (W @ A[t, :]) / np.sum(W)

#         if 0 > A[t+1, x]:
#             A[t+1, x] = 0
#         if 1 < A[t+1, x]:
#             A[t+1, x] = 1
        
```
