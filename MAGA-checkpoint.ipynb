```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation
```

```python
# This code runs an opinion model simulation. It uses the basic model:

    # For an opinion x0 at t=0, and an opinion proximity R,
    # for any opinion x, x is in the set X0 iff |x - x0| < R. 
    # x1 at t=1 is then x1 = mean(all x in X0).
    
# The model contains:

    # advertisers = agents whose opinion remains forcibly constant but still influence, 
                    # the degree to which they influence can be controlled by their preset "charisma" value
                    # (this value determines the number of times they are added in to the mean calculation of agents R-close)
    
    # polling = opinions are discretised at each timestep (i.e. agents can only be influenced seeing a discrete poll)
    
    # brownian motion = randomness added to each agents opinion at each timestep
    
    # advertiser drift = predetermined movement of advertiser opnions with time
    
```

```python

# set number of agents and time increments
Nx = 200
Nt = 15

# end and start times
t0 = 0
t1 = 1.

# Varying radius's of influence
R_list = [0.025, 0.1, 0.2, 0.25]
R = R_list[3]

# Random, uniform, normal and normal-double-hump initial opinion distributions
rand_Ics = np.sort(np.random.random(Nx))
uni_Ics = np.linspace(0., 1., Nx)
norm_Ics = np.sort(np.random.normal(loc=0.5, scale=0.2, size=Nx))

twohumps_Ics = np.concatenate((np.sort(np.random.normal(loc=0.2, scale=0.1, size=int(Nx/2))), 
                               np.sort(np.random.normal(loc=0.8, scale=0.1, size=int(Nx/2)))), axis=0) 

random_start = True
normal_start = False
uniform_start = False
twohumps_start = False

# other parameters relevant to the project
adverts = False
advertisers = [0.5]
charisma = [75]
advertiser_drift = False
drift = [2] 
advertiser_sine_wave = False

polling = False

brownian_motion = False
h = .08 # the variance of the brownian motion

radius_weights = False

# dont worry about this -- yet to be implemented
T = np.linspace(0, Nt, Nt)
T_scaled = (T/Nt)*2*np.pi
```

```python
# initialising opinion matrices (agents vs time)

if twohumps_start:
    twohumps_Ics = np.array([x for x in twohumps_Ics if 0 < x < 1])
    A = np.zeros([Nt, len(twohumps_Ics)])
    A[0, :] = twohumps_Ics

if random_start:
    A = np.zeros([Nt, Nx])
    A[0, :] = rand_Ics
    
elif uniform_start:
    A = np.zeros([Nt, Nx])
    A[0, :] = uni_Ics
    
elif normal_start:
    norm_Ics = np.array([x for x in norm_Ics if 0 < x < 1])
    A = np.zeros([Nt, len(norm_Ics)])
    A[0, :] = norm_Ics
    
ad_pos = []

if adverts:
# ------ Add advertisers into the system
### Add preset values to the initial array to be kept constant and purely influence others
    for i in range(len(advertisers)):
        A_new = np.hstack((A, np.atleast_2d(np.zeros(Nt)).T))
        A_new[0, -1] = advertisers[i]
        A = A_new

    A[0, :] = np.sort(A[0, :]) #resort the initial opinions with adverts slotted in


    ### find and record the matrix rows representing advertisers
    for stance in advertisers:
        pos = np.argwhere(A[0, :] == stance)[0][0]
        ad_pos.append(pos)
        
ad_pos = np.array(ad_pos)
# ------------------------------------------------
            
##### We update opinions based on 
for t in range(A.shape[0]-1):

    U = A[t, :]
    
    # round opinions (take a poll)
    if polling:
        U_poll = np.round(A[t, :], 1)
    
    # find advert values
    if adverts:
        ads = U[ad_pos]
        if polling:
            ads_poll = U_poll[ad_pos]
    
    for x in range(A.shape[1]):
    
        A[t, x] = U[x]
        
        if x not in ad_pos:
            
            # add random fluctuation
            if brownian_motion:
                random_increment = np.random.normal(0.0, 1.0 * h, 1)*np.sqrt(1/(5*Nt))
                A[t, x] += random_increment
        
            # find R-close opinions
            close_agents = U[abs(U - A[t, x]) < R]
            
            if polling:
                close_polls = U_poll[abs(U_poll - A[t, x]) < R]
                
                if adverts:
                    new_op = 0
                    # find positions of close ads
                    close_ads_poll = ads_poll[abs(ads_poll - A[t, x]) < R]
                    
                    ### identify the opinions values of these ads
                    if len(close_ads_poll) > 0:
                        close_ad_num = []
                        for a in close_ads_poll:
                            for i in ad_pos:
                                if U[i] == a: 
                                    which_ad = np.where(ad_pos == i)[0][0]
                            close_ad_num.append(which_ad)

                        new_op += np.sum(close_polls)
                        
                        ## include adverts into the mean *charisma* times
                        total_charisma = 0

                        for i in range(len(close_ads_poll)):
                            ad_num = close_ad_num[i]
                            new_op += charisma[ad_num] * advertisers[ad_num]
                            total_charisma += abs(charisma[ad_num])

                        new_op = new_op / (len(close_polls) + total_charisma)

                        A[t+1, x] = new_op
                        
                    else:
                        A[t+1, x] = np.mean(close_polls)
                else:     
                    A[t+1, x] = np.mean(close_polls)
                    
                        
            else:

                if adverts:

                    new_op = 0 
                    ### identify close adverts...
                    close_ads = ads[abs(ads - A[t, x]) < R]

                    if len(close_ads) > 0:
                       ### ...and their values
                        close_ad_num = []
                        for a in close_ads:
                            for i in ad_pos:
                                if U[i] == a:
                                    which_ad = np.where(ad_pos == i)[0][0]
                            close_ad_num.append(which_ad)
                        
                        new_op += np.sum(close_agents)

                        total_charisma = 0
                        
                        ### add advertiser opinions into the mean of each opinion x_t "charisma" times
                        for i in range(len(close_ads)):
                            ad_num = close_ad_num[i]
                            new_op += charisma[ad_num] * advertisers[ad_num]
                            total_charisma += abs(charisma[ad_num])

                        new_op = new_op / (len(close_agents) + total_charisma)

                        A[t+1, x] = new_op

                    else:
                        A[t+1, x] = np.mean(close_agents)
                else:     
                    A[t+1, x] = np.mean(close_agents)

        else:
            
            # Update advertiser opinions based on predetermined movement 
            if advertiser_drift:
                ad_index = np.where(ad_pos == x)[0][0]
                A[t, x] += (drift[ad_index] / Nt) * A[t, x]
             
            # advertisers opinions forming a sine wave in time (popularity politics)
            if advertiser_sine_wave:
                A[t+1, x] = 0.5 + 0.4 * np.sin(0.315 * T[t+1])
                
            else:
                A[t+1, x] = A[t, x]
            
        ## tidying up opinion edges
        if 0 > A[t+1, x]:
            A[t+1, x] = 0
        if 1 < A[t+1, x]:
            A[t+1, x] = 1


### a final random fluctuation
if brownian_motion: 
               
    for x in range(A.shape[1]):        
        random_increment = np.random.normal(0.0, 1.0 * h, 1)*np.sqrt(1/(5*Nt))
        A[-1, x] += random_increment

## a final ad movement
if advertiser_drift:
    for x in range(Nx):
        if x in ad_pos:
            ad_index = np.where(ad_pos == x)[0][0]
            A[-1, x] += (drift[ad_index] / Nt) * A[t, x]
```

```python
T = np.linspace(0, Nt, Nt)

fig, ax = plt.subplots(figsize=(8,10))
ax.set_xticks(range(Nt+1))
ax.set_yticks(np.linspace(0., 1., 11))
ax.tick_params(right=True, labelright=True)
ax.set_title(fr"Cluster Arrangement: $R = {R}$", fontdict={'fontsize': 26, 'fontweight': 'roman'})
ax.set_xlabel("Timestep", fontdict={'fontsize': 16, 'fontweight': 'roman'})
ax.set_ylabel("Opinion stance (0 = totally against, 1 = totally for)", fontdict={'fontsize': 16, 'fontweight': 'roman'})
ax.axis([0, Nt, 0, 1])

for t in range(A.shape[1]):
    ax.plot(T, A[:, t])

# # ax.plot(T, A[:, ad_pos[3]], 'om', label="Charisma = 500")
# # ax.plot(T, A[:, ad_pos[2]], 'oc', label="Charisma = 100")
# # ax.plot(T, A[:, ad_pos[1]], 'or', label="Charisma = 10")
# ax.plot(T, A[:, ad_pos[0]], 'ok', label=f"Charisma = {charisma[0]}")

# # leg = plt.legend(loc="best", fontsize="x-large")
plt.show()
# plt.savefig(f"clusters_R={R}.png")
```
